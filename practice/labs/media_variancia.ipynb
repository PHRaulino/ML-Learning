```python
import math
import pandas as pd
```

```python
pessoas = pd.read_csv('dados.csv', sep=';')
pessoas
```

```python
pessoas_resumo = pessoas.groupby('pessoa').count().rename(columns={'valor': 'contagem'}).reset_index().copy()
pessoas_media = pessoas.groupby('pessoa').mean().rename(columns={'valor': 'media'}).reset_index().copy()
pessoas = pessoas.merge(pessoas_media, how='left')
pessoas['diff_media_valor_quadrado'] = (pessoas.valor - pessoas.media) ** 2
soma_varianca = pessoas[['pessoa', 'diff_media_valor_quadrado']].groupby('pessoa').sum().reset_index().copy()

pessoas
```

```python
pessoas.groupby('pessoa').agg(['count', 'sum', 'mean', 'var', 'std'])
```

```python
pessoas_resumo = pessoas_resumo.merge(pessoas_media, how='left')
pessoas_resumo = pessoas_resumo.merge(soma_varianca, how='left')
pessoas_resumo['varianca'] = pessoas_resumo.diff_media_valor_quadrado / pessoas_resumo.contagem
pessoas_resumo['desvio_padrao'] = pessoas_resumo.varianca.apply(math.sqrt).round(2)
pessoas_resumo
```
