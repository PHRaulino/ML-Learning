```python
# ✅ SETUP — conexão com o banco
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import psycopg2
from sqlalchemy import create_engine

# Ajuste host/porta/credenciais conforme seu ambiente (docs/database.md)
engine = create_engine('postgresql://user:password@raspberry-pi-host:5432/fintrack')

print('Conexão configurada.')
```

```python
# ✅ SETUP — carrega despesas de conta (expenses)
query = """
SELECT
    e.transaction_date,
    e.amount,
    e.description,
    c.name AS categoria,
    sc.name AS subcategoria,
    a.name AS conta
FROM expenses e
JOIN categories c ON c.id = e.category_id
LEFT JOIN sub_categories sc ON sc.id = e.sub_category_id
JOIN accounts a ON a.id = e.account_id
WHERE e.deleted_at IS NULL
  AND e.transaction_status = 'PAID'
ORDER BY e.transaction_date
"""

df = pd.read_sql(query, engine)
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['amount'] = df['amount'].astype(float)

print(f'Linhas: {len(df)}')
df.head(10)
```
